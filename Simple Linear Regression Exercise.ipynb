## Import the relevant libraries

```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as pit
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

## Load the data

```python
data = pd.read_csv('real_estate_price_size.csv')
print(data)
```

```python
print(data.describe())
```

## Create the regression

### Declare the dependent and the independent variables

```python
y = data['price'] #dependant
x1 = data['size'] #independant
```

### Explore the data

```python
pit.scatter(x1,y)
pit.xlabel('size',fontsize=20)
pit.ylabel('price',fontsize=20)
pit.show()
```

### Regression itself

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit() #it containts the output of ordinary least squares regression (MCO in spanish, MÃ­nimos cuadrados ordinarios)
print(results.summary()) # we show us the result of the regression
```

### Plot the regression line on the initial scatter

```python
#we print a regression line on the last graphic
pit.scatter(x1,y)
yhat = 223.1787*x1 + 1.019e+05
fig = pit.plot(x1,yhat,lw=4,c = 'orange', label="regression line")
pit.xlabel('SAT',fontsize=20)
pit.ylabel('GPA',fontsize=20)
pit.show()
```
