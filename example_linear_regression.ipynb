```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pit
import statsmodels.api as sm
```

```python
#Load the data
```

```python
data = pd.read_csv('1.01. Simple linear regression.csv')
print(data)
```

```python
#show the most utils statistics for each column
print(data.describe())
```

```python
# We have a sample of university students (84)
#STA = Lectura crítica + matemáticas + escritura
#GPA = Promedio universitario de calificaciones al momento de graduarse
# We'll create a linear regression, which predicts a student's GPA based on the STA
```

```python
#define the dependant and the independant variables
y = data['GPA'] #dependant
x1 = data['SAT'] #independant
```

```python
#explore the data
pit.scatter(x1,y)
pit.xlabel('SAT',fontsize=20)
pit.ylabel('GPA',fontsize=20)
pit.show()
```

```python
#each point is a different student
#We can see a strong relationship between SAT and GPA
#In general,The higher a student's score on the STA, the higher his or her GPA will be.
#Regression itself
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit() #it containts the output of ordinary least squares regression (MCO in spanish, Mínimos cuadrados ordinarios)
print(results.summary()) # we show us the result of the regression
```

```python
#we print a regression line on the last graphic
pit.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
fig = pit.plot(x1,yhat,lw=4,c = 'orange', label="regression line")
pit.xlabel('SAT',fontsize=20)
pit.ylabel('GPA',fontsize=20)
pit.show()
```
